### Designation

```python
df['Designation'].unique()
```

```python
df['Designation'].value_counts()
```

```python
l=[]
for i in df['Designation']:
    if ('senior' in i and 'engineer' not in i):
        l.append('senior')
    elif('trainee'in i and 'engineer' not in i):
        l.append('trainee')
    elif('engineer' in i and 'senior' not in i):
        l.append('engineer')
    elif('associate' in i and 'senior' not in i):
        l.append('associate')
    elif('developer' in i and 'senior' not in i):
        l.append('developer')
    elif('manager' in i and 'senior' not in i):
        l.append('manager')
    elif('analyst' in i):
        l.append('analyst')
    elif('consultant' in i):
        l.append('consultant')
    elif('executive' in i):
        l.append('executive')
    elif('designer' in i):
        l.append('designer')
    else:
        l.append('others')
```

```python
df['Designation'] = l
df['Designation'].value_counts()
```

```python
plt.figure(figsize=(14,6))
sns.countplot(x=df['Designation'])
plt.show()
```

### observation
- Most peples designation are an engineer
